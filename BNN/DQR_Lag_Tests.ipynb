```
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```
lag_lst = np.arange(2, 46, 2)
mse_i = np.array(pd.read_csv('Lag Tests/Results/mse_i.csv', header=None))[:,:22]
mse_j = np.array(pd.read_csv('Lag Tests/Results/mse_j.csv', header=None))[:,:22]
icp_i = np.array(pd.read_csv('Lag Tests/Results/icp_i.csv', header=None))[:,:22]
icp_j = np.array(pd.read_csv('Lag Tests/Results/icp_j.csv', header=None))[:,:22]
mil_i = np.array(pd.read_csv('Lag Tests/Results/mil_i.csv', header=None))[:,:22]
mil_j = np.array(pd.read_csv('Lag Tests/Results/mil_j.csv', header=None))[:,:22]
time_i = np.array(pd.read_csv('Lag Tests/Results/time_mean.csv', header=None))[:,:22]
time_j = np.array(pd.read_csv('Lag Tests/Results/time_joint.csv', header=None))[:,:22]
time_q = np.array(pd.read_csv('Lag Tests/Results/time_quan.csv', header=None))[:,:22]
```

```
tradeoff_i = np.abs(icp_i - 0.90)*mil_i
tradeoff_j = np.abs(icp_j - 0.90)*mil_j
```

```
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4))
l1 = ax1.plot(lag_lst, mse_i[1], 'bo')
l2 = ax1.plot(lag_lst, mse_j[1], 'ro')
ax1.set_xlabel('Lags')
ax1.set_ylabel('MSE')

ax2.plot(lag_lst, tradeoff_i[4], 'bo')
ax2.plot(lag_lst, tradeoff_j[4], 'ro')
ax2.set_xlabel('Lags')
ax2.set_ylabel('Tradeoff')

ax3.plot(lag_lst, time_i[4] + time_q[4], 'bo')
ax3.plot(lag_lst, time_j[4], 'ro')
ax3.set_xlabel('Lags')
ax3.set_ylabel('Training Time (Seconds)')

ax1.legend([l1, l2],   
          labels=['Independent', 'Joint'],  
          loc="upper right", 
          borderaxespad=0.5,   
          frameon=True)
plt.tight_layout()
```
