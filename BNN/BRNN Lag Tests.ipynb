```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
lag_lst = np.arange(2, 40, 4)
mse_b = np.array(pd.read_csv('mse_basic.csv'))[:,1:]
mse_l = np.array(pd.read_csv('mse_tied.csv'))[:,1:]
icp_b = np.array(pd.read_csv('icp_basic.csv'))[:,1:]
icp_l = np.array(pd.read_csv('icp_tied.csv'))[:,1:]
mil_b = np.array(pd.read_csv('mil_basic.csv'))[:,1:]
mil_l = np.array(pd.read_csv('mil_tied.csv'))[:,1:]
time_b = np.array(pd.read_csv('time_basic.csv'))[:,1:]
time_l = np.array(pd.read_csv('time_tied.csv'))[:,1:]
```

```python
tradeoff_b = np.abs(icp_b - 0.90)*mil_b
tradeoff_l = np.abs(icp_l - 0.90)*mil_l
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4))
l1 = ax1.plot(lag_lst, mse_b.T, 'bo')
l2 = ax1.plot(lag_lst, mse_l.T, 'ro')
ax1.set_xlabel('Lags')
ax1.set_ylabel('MSE')

ax2.plot(lag_lst, tradeoff_b.T, 'bo')
ax2.plot(lag_lst, tradeoff_l.T, 'ro')
ax2.set_xlabel('Lags')
ax2.set_ylabel('Tradeoff')

ax3.plot(lag_lst, time_b.T, 'bo')
ax3.plot(lag_lst, time_l.T, 'ro')
ax3.set_xlabel('Lags')
ax3.set_ylabel('Training Time (Seconds)')

ax2.legend([l1, l2],   
          labels=['Basic', 'Tied LSTM'],  
          loc="upper left", 
          borderaxespad=0.5,   
          frameon=True)
plt.tight_layout()
```
